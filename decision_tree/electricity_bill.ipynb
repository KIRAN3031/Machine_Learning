```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

```python
# units consumed
X = np.array([50, 120, 250, 400, 600]).reshape(-1,  1)
```

```python
# electricity bill amount
y = np.array([100, 220, 450, 700, 1000])
```

```python
dt_regressor = DecisionTreeRegressor(max_depth=3, random_state=42)
dt_regressor.fit(X, y)
```

```python
X_grid = np.linspace(0, 650, 100).reshape(-1, 1)
y_pred = dt_regressor.predict(X_grid)
```

```python

acc = dt_regressor.score(X, y)
print(f"Model Accuracy: {acc*100:.2f}%")
```

```python
plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(X_grid, y_pred, color='blue', label='Decision Tree Prediction')
plt.title('Electricity Bill Prediction using Decision Tree Regression')
plt.xlabel('Units Consumed')
plt.ylabel('Electricity Bill Amount')
plt.legend()
plt.grid(True)
plt.show()
```
