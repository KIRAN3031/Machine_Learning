```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.arange(1, 21).reshape(-1, 1)
y = np.array([10, 12, 15, 18, 20, 25, 30, 28, 35, 40,
              45, 42, 50, 55, 60, 58, 65, 70, 72, 75])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train, y_train)
```

```python
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)
```

```python
# Predictions
y_pred_dt = dt_regressor.predict(X_test)
y_pred_rf = rf_regressor.predict(X_test)
```

```python
# Metrics
from sklearn.metrics import mean_squared_error
print("Decision Tree Regressor Predictions:", y_pred_dt)
print("Decision Tree Regressor R^2 Score:", dt_regressor.score(X_test, y_test))
print("Decision Tree Regressor Mean Squared Error:", mean_squared_error(y_test, y_pred_dt))
print("\n\n\n")
print("Random Forest Regressor Predictions:", y_pred_rf)
print("Random Forest Regressor R^2 Score:", rf_regressor.score(X_test, y_test))
print("Random Forest Regressor Mean Squared Error:", mean_squared_error(y_test, y_pred_rf))
```

```python
X_grid = np.linspace(min(X), max(X), 100).reshape(-1, 1)
y_pred_grid_dt = dt_regressor.predict(X_grid)
y_pred_grid_rf = rf_regressor.predict(X_grid)
```

```python
# visualization
plt.figure(figsize=(12, 6))
plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(X_grid, y_pred_grid_dt, color='blue', label='Decision Tree Regressor', linewidth=2)
plt.plot(X_grid, y_pred_grid_rf, color='green', label='Random Forest Regressor', linewidth=2)
plt.title('House Cost Prediction using Decision Tree and Random Forest Regressors')
plt.xlabel('Feature (e.g., Size in 1000 sqft)')
plt.ylabel('House Cost (in $1000s)')
plt.legend()
plt.grid(True)
plt.show()
```
