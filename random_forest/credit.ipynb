## Credit Card Farud Detection

## importing the neccessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
```

## Loading the data set

```python
import kagglehub
import pandas as pd
import os

path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")
print("Files:", os.listdir(path))  # Shows ['creditcard.csv']
```

```python

df = pd.read_csv(os.path.join(path, 'creditcard.csv'))
```

## Feature separation

```python
X = df.drop("Class",axis=1)
y= df.Class
```

## Train Test split

```python
# splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42, stratify=y)
```

## Feature scaling or transformation

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Model Building (Decision Tree)

```python
# training the Decision Tree Classifier
classifier1 = DecisionTreeClassifier(
    criterion="entropy", random_state=42 , max_depth=4
)
classifier1.fit(X_train, y_train)
```

## Training Score

```python
# model score
score = classifier1.score(X_train, y_train)
print(f"Model Accuracy: {score * 100:.2f}%")
```

## Testing Score

```python
# model evaluation
y_pred = classifier1.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

## Model Building (Random Forest)

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(
    n_estimators=100,
    criterion="entropy",
    max_depth=4,
    random_state=42
)
classifier.fit(X_train, y_train)
```

## Training Score

```python
# model score
score = classifier.score(X_train, y_train)
print(f"Model Accuracy: {score * 100:.2f}%")
```

## Testing Score

```python
# model evaluation
y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

## Confusion Matrix

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
```

## Classification Report

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

## Tree Plot Random Forest

```python
plt.figure(figsize=(12,8))
plot_tree(classifier.estimators_[0],
          feature_names=X.columns.tolist(),
          class_names=['Not Fraud', 'Fraud'],
          filled=True,
          rounded=True)
plt.title("First Random Forest Tree")
plt.show()
```
